```python
class DiskInfo(object) :
    def __init__(self,k1,k2,k3,
                      z1,z2,z3) :
        self.k1 = k1
        self.k2 = k2
        self.k3 = k3
        self.k4 = None
        self.z1 = z1
        self.z2 = z2
        self.z3 = z3
        self.z4 = None

    def ks(self) :
        return array([self.k1,self.k2,self.k3],dtype=float)

    def zs(self) :
        return array([self.z1,self.z2,self.z3],dtype=complex)

def take(n, iterable):
    "Return first n items of the iterable as a list"
    return islice(iterable, n)

def disk_generator() :
    inf = complex('inf')
    disk = DiskInfo(k1=0,k2=0,k3=1,
                    z1=inf,z2=inf*1.0j,z3=0+0j)
    disk_list = [disk]
    while True :
        disk = disk_list[0]
        ks = disk.ks()
        zs = disk.zs()
        indices = argsort(ks)
        k1, k2, k3 = ks[indices]
        z1, z2, z3 = zs[indices]
        num_lines = sum(ks==0)
        
        if num_lines == 2 :
            k4 = k3 * (3 + sqrt(8))
            z4 = z3 + (sqrt(2)/2)*(1+1j)*(1/k3 + 1/k4)
        else :
            if num_lines == 0 :
                zk1 = z1*k1
            else :
                zk1 = 1.0 if z1==complex('inf') else 1.0j
            zk2 = z2*k2
            zk3 = z3*k3
            k4 = k1 + k2 + k3 + 2*sqrt(k1*k2 + k1*k3 + k2*k3)
            z4 = (zk1 + zk2 + zk3 + 2*sqrt(zk1*zk2+zk1*zk3+zk2*zk3))/k4

        disk.k4 = k4
        disk.z4 = z4
        yield disk
        
        disk_list.append(DiskInfo(k1=k1,k2=k2,k3=k4,
                                  z1=z1,z2=z2,z3=z4))
        disk_list.append(DiskInfo(k1=k1,k2=k4,k3=k3,
                                  z1=z1,z2=z4,z3=z3))
        disk_list.append(DiskInfo(k1=k4,k2=k2,k3=k3,
                                  z1=z4,z2=z2,z3=z3))
        del disk_list[0]            
```

```python
from itertools import islice

fig = figure(figsize=(10,10))
ax = fig.gca()
ax.add_artist(Circle((0,0),1,color='0.75'))

level = 6
ndisks = sum(map(lambda xx:3**xx,range(level+1)))

area = np.pi
for ndisk, disk in enumerate(take(ndisks,disk_generator())) :
    radius = 1/disk.k4
    z = disk.z4
    x,y = z.real, z.imag
    color = str(1.0/(1+int(np.log(3*(ndisk+1))/np.log(3))))
    for xy in [(x,y),(-x,y),(x,-y),(-x,-y)] :
        ax.add_artist(Circle(xy,radius,color=color))
        area += np.pi * radius**2
xlim([-1,+1])
ylim([-1,+1])
print(area)
```
