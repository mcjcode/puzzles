```python
import numpy as np
```

```python
from matplotlib import pyplot as plt
```

```python
def proposed_expected_time(n,p):
    if p == 1:
        return n
    else:
        pn = p**n
        q = 1-p
        return (2*(1-pn)/q-n*pn)/((1+n*q)*pn)
```

```python
fig = plt.figure(figsize=(13,11))
ax = fig.add_subplot(2, 1, 1)
ax.set_yscale('log')
for n in range(1,6):
    #line, = ax.plot(a, color='blue', lw=2)
    ps = np.linspace(0.01,0.99,1000)
    ts = np.array(list(map(lambda p: proposed_expected_time(n,p)*(p/n), ps)))
    ax.plot(ps,ts)
```

```python
proposed_expected_time(10,0.999)
```
