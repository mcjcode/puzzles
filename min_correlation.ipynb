```python
%pylab inline
import numpy as np
from fractions import Fraction
```

```python
def frac_formatter(frac) :
    n = frac.numerator
    d = frac.denominator
    st = '%d' % n if (d==1) else '%d/%d' % (n,d)
    return '%6s' % st

def printmat(mat) :
    print(np.array2string(mat,
                          formatter = {'all':frac_formatter},
                          max_line_width=200))
```

```python
def pascal(n) :
    k = 0
    result = np.array([1])
    while k < n :
        result = np.array(result.tolist() + [0]) + np.array([0] + result.tolist())
        k += 1
    return result
```

```python
def make_matrix(n,k=None) :
    """
    Write down the constraints on b0,...,bn
    that say that they are a probability
    distribution, that the coins are fair,
    and that all correlations up to the (k-1)st
    order correlations are zero.
    """
    if k is None :
        k = n-2
    row = pascal(n-k)
    a = np.zeros((k+1,k+4),dtype=Fraction)
    for i in range(k+1) :
        a[i,i:(i+row.shape[0])] = row
        a[i,n+1] = Fraction(1,2**(n-2))
    a[n-2,n+1] = Fraction(1,2**(n-2))
    return a
```

```python
a=make_matrix(7,5)
```

```python
printmat(a)
```

```python
def reduce_matrix(a) :
    n = a.shape[0]+1
    for i in range(n-2,-1,-1) :
        for j in range(i+1,n-1) :
            a[j,:] = a[j,:] - a[j,i+2] * a[i,:]
        a[i,i+2] = 0
    return a[:,[0,1,-1]]
```

```python
a = reduce_matrix(a)
printmat(a)
```

```python
n = 7
a = make_matrix(n)
a = reduce_matrix(a)
fig = figure(figsize=(10,10))
for i in range(a.shape[0]) :
    xs = np.array([0,a[i,2]/a[i,0]])
    ys = np.array([a[i,2]/a[i,1],0])
    color = 'r' if (i%2) else 'b'
    plot(xs,ys,color=color)
xlocs, xlabs = xticks([1./2**(n-1),1./2**(n-2)],['$1/2^{%d}$'%(n-1),'$1/2^{%d}$'%(n-2)],size=20)
ylocs, ylabs = yticks([1./2**(n),1./2**(n-1)],  ['$1/2^{%d}$'%(n),'$1/2^{%d}$'%(n-1)],size=20)
xlabel('$b_0$',size=24)
ylabel('$b_1$',size=24)
grid(True)
```

So it looks like if we have $n\gt\gt 0$ coins, the minimum value of $b_0+b_1$ is approximately $1/2^{n-1}$, and the minimum value of $\rho_{n-1}$ is then approximately $2^{n-1}a_{n-1}-1 \sim 0$.
